```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.style.use('seaborn-whitegrid')

import numpy as np
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 1000)
ax.axis([-1, 11, -1.5, 1.5]);
ax.plot(x, np.sin(x));
```

```python
#alterando o tamanho da fonte

plt.rcParams.update({'font.size':15})

fig = plt.figure()
ax = plt.axes()

#linha solida, cor especificada pelo nome
plt.plot(x, np.sin(x - 0),  color = 'blue', linestyle = 'solid', label = 'azul')

#definindo a cor com o nome curto
plt.plot(x, np.sin(x - 1),  color = 'g', linestyle = 'dashed', label = 'verde')

#escala de cinza entre 0 e 1 
plt.plot(x, np.sin(x - 2),  color = '0.75', linestyle = 'dashdot', label = 'cinza')

#definindo a cor pelo RGB
plt.plot(x, np.sin(x - 3),  color = '#ff0000', linestyle = 'dotted', label = 'vermelho')

#limites dos eixos
plt.axis([-1, 11, -1.5, 1.5]);

#titulo

plt.title("Exemplo de um gráfico", color='green')

#legendas
plt.legend(loc = 'lower left')
```

```python
#dados discretos

x = np.linspace(0, 10, 50)
dy = 0.8
y= np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr = dy, fmt ='o', color='red', ecolor='lightgray', elinewidth=3, capsize=0);
```

```python
#Customizando o grafico

print(plt.style.available[:6])

fig = plt.figure(figsize=(12,8))

for i in range(6):
    #Adicionando suplots
    fig.add_subplot(3,2,i+1)
    plt.style.use(plt.style.available[i])

    plt.plot(x,y)
    #escrevendo o subplot
    plt.text(s=plt.style.available[i], x=5, y=2, color='green', ha='center', va='center')
    
```

```python
x = np.random.randn(1000)

plt.style.use('classic')
fig=plt.figure(figsize=(5,3))
ax = plt.axes(facecolor='#E6E6E6')
# Display ticks underneath the axis
ax.set_axisbelow(True)
# White frame
plt.grid(color='gray', linestyle='solid')

# Hide the frame
for spine in ax.spines.values():    
    spine.set_visible(False)
    
# Hide the markers at the top and the right
ax.xaxis.tick_bottom()
ax.yaxis.tick_left()

# We can personalise the markers, and rotate them
marqueurs = [160, -2, -1, 0, 1, 2, 3]
xtick_labels = ['A', 'B', 'C', 'D', 'E', 'F']
plt.xticks(marqueurs, xtick_labels, rotation=30)

# Change the color of markers
ax.tick_params(colors='gray', direction='out')
for tick in ax.get_xticklabels():    
    tick.set_color('gray')
for tick in ax.get_yticklabels():    
    tick.set_color('gray')    
    
# Change the color of the edges
ax.hist(x, edgecolor='#000000', color='#51e0a7');
```

```python
#import matplotlib
#import matplotlib.pyplot as plt
#import numpy as np

labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels))  # the label locations
width = 0.40  # the width of the bars

fig, ax = plt.subplots()

#definindo as barras ("comprimento","altura", etc... )
rects1 = ax.bar(x - width/2, men_means, width, label='Men', color='#2f56f5')
rects2 = ax.bar(x + width/2, women_means, width, label='Women', color='#f26f66')


ax.axis([-1, 5, 0, 40]);

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender', loc='center', color='#0525a6')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')



autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
