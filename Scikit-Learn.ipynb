```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as skl

sns.set()

hubble = pd.read_csv("hubble_data.csv")
hubble.describe

#x = np.linspace(0, 10, 500)
#y = np.random.randn(500)
#plt.plot(x,y)
```

```python
from sklearn.linear_model import LinearRegression

X = hubble.distance.values.reshape(-1,1)
Y = hubble.recession_velocity

lr = LinearRegression()
lr.fit(X, Y)

BIGGER_SIZE = 15

plt.rc('font', size=BIGGER_SIZE) 
plt.rc('axes', titlesize=BIGGER_SIZE) 
plt.rc('axes', labelsize=BIGGER_SIZE) 
plt.rc('xtick', labelsize=BIGGER_SIZE)
plt.rc('ytick', labelsize=BIGGER_SIZE) # fontsize of the tick labels 
plt.rc('legend', fontsize=BIGGER_SIZE) # legend fontsize 
plt.rc('figure', titlesize=BIGGER_SIZE) # fontsize of the figure title
plt.figure()
plt.title("Data from Hubble")
plt.scatter(X, Y, color='black', label="data")
plt.plot(X, lr.predict(X), color='red', linewidth=3, label="fitted line") 
plt.xlabel("Distance")
plt.ylabel("Recession velocity")
plt.legend()
plt.show()
```
