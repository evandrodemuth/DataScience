```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()

arvores = pd.read_csv("p2-arbres-en.csv",sep=";")


arvores["qty"] = 1
#arvores.describe
arvores.info()

```

```python
arvores.describe(include="all")
```

```python
# Comentar que irei realizar a preparação do Dataframe eliminando os NAN e inferindo valores nas variaveis categoricas,

#elimina columnas com valores nulos
#arvores = arvores.dropna(axis="columns")
arvores.tree_specie.head(200)
```

```python
#arvores.head(4)
```

```python

# Seleciona somente a coluna de especie e a de quantidade
dados = arvores[["tree_specie","qty"]]

# Agrupa por Especie de arvore
gb = dados.groupby("tree_specie")

#reset_index para converter o groupby em dataframe
grouped = gb.count().reset_index()

# 1) ordenar ascendente
# 2) limitar top 10 arvores 
shortGroup = grouped.sort_values("qty", ascending = False).head(10).reset_index()

# Elimina coluna index indesejada
shortGroup = shortGroup.drop(columns=['index'])
```

```python
shortGroup
```

```python
#

```

```python
sns.set(style="whitegrid")

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(15, 6))
sns.barplot(x="tree_specie", y="qty", data=shortGroup, label="Total", color="b")
ax.set_ylabel("Quantity", fontsize=15)
ax.set_xlabel("Tree Specie", fontsize=15)
ax.set_title("Top 10 tree species of Paris" ,fontsize=20)
```
